```r
library(cluster)

nihao_noagn <- read.csv("NIHAO_NoAGN.csv")
nihao_agn <- read.csv("NIHAO_AGN.csv")
nihao_uhd <- read.csv("NIHAO_UHD.csv")
nihao_n80 <- read.csv("NIHAO_n80.csv")

sdss <- read.csv("SDSS.csv")
```

### PCA
We can do a simple principal component analysis to see whether the intrinsic dimensionality of the latent variable  space of the SDSS and of NIHAO are the same. There was [some debate](https://www.nature.com/articles/nature07366) in the past on galaxies being 'simpler' than expected (i.e. being described by fewer latent variables). Is the number of independent latent variables in simulation and in reality the same? This is an important check.

##### NIHAO NO AGN (blue) VS SDSS (brown)

```r
prcomp_nihao_noagn <- prcomp(nihao_noagn[,3:130])
prcomp_nihao_agn <- prcomp(nihao_agn[,3:130])
prcomp_nihao_uhd <- prcomp(nihao_uhd[,3:130])
prcomp_nihao_n80 <- prcomp(nihao_n80[,3:130])
prcomp_sdss <- prcomp(sdss[,3:130])
```

```r
nihao_noagn_color <- "#81D6E3"
nihao_agn_color <- "#8075FF"
nihao_uhd_color <- "#21295C" 
nihao_n80_color <- "#4CB5AE"
sdss_color <- "#AF9164"
thin_line_size <- 0.5
symbol_size <- 0.8
```

```r
var_explained <- function(prcomp_model) prcomp_model$sdev^2 / sum(prcomp_model$sdev^2)

plot(1:12, var_explained(prcomp_nihao_noagn)[1:12], pch = 16, cex = symbol_size, col = nihao_noagn_color, ylab = 'Explained variance', xlab = "PC") 
lines(1:12, var_explained(prcomp_nihao_noagn)[1:12], lwd = thin_line_size, col = nihao_noagn_color)
points(1:12, var_explained(prcomp_nihao_agn)[1:12], pch = 16, cex = symbol_size, col = nihao_agn_color)
lines(1:12, var_explained(prcomp_nihao_agn)[1:12], lwd = thin_line_size, col = nihao_agn_color)
points(1:12, var_explained(prcomp_nihao_uhd)[1:12], pch = 16, cex = symbol_size, col = nihao_uhd_color)
lines(1:12, var_explained(prcomp_nihao_uhd)[1:12], lwd = thin_line_size, col = nihao_uhd_color)
points(1:12, var_explained(prcomp_nihao_n80)[1:12], pch = 16, cex = symbol_size, col = nihao_n80_color) 
lines(1:12, var_explained(prcomp_nihao_n80)[1:12], lwd = thin_line_size, col = nihao_n80_color)
points(1:12, var_explained(prcomp_sdss)[1:12], pch = 16, cex = symbol_size, col = sdss_color)
lines(1:12, var_explained(prcomp_sdss)[1:12], lwd = thin_line_size, col = sdss_color)


plot(1:128, log10(var_explained(prcomp_nihao_noagn)), pch = 16, cex = symbol_size, 
     col = nihao_noagn_color, ylab = 'Log explained variance', xlab = "PC", ylim = range(log10(var_explained(prcomp_nihao_n80)))) 
lines(1:128, log10(var_explained(prcomp_nihao_noagn)), lwd = thin_line_size, col = nihao_noagn_color)

points(1:128, log10(var_explained(prcomp_nihao_agn)), pch = 16, cex = symbol_size, col = nihao_agn_color) 
lines(1:128, log10(var_explained(prcomp_nihao_agn)), lwd = thin_line_size, col = nihao_agn_color)

points(1:120, log10(var_explained(prcomp_nihao_uhd)), pch = 16, cex = symbol_size, col = nihao_uhd_color) 
lines(1:120, log10(var_explained(prcomp_nihao_uhd)), lwd = thin_line_size, col = nihao_uhd_color)

points(1:128, log10(var_explained(prcomp_nihao_n80)), pch = 16, cex = symbol_size, col = nihao_n80_color) 
lines(1:128, log10(var_explained(prcomp_nihao_n80)), lwd = thin_line_size, col = nihao_n80_color)

points(1:128, log10(var_explained(prcomp_sdss)), pch = 16, cex = symbol_size, col = sdss_color) 
lines(1:128, log10(var_explained(prcomp_sdss)), lwd = thin_line_size, col = sdss_color)
```

Interestingly, the SDSS sample behaves differently from the NIHAO No AGN, NIHAO AGN, and NIHAO n80 samples (in different shades of light blue): the former has the first PC explaining 1/4 of the total variance with the following PCs dropping quickly below 10%, while the latter have the first few PCs contributing more equally. Moving towards the least important PCs, which explain residual variance (noise), we see that these lose importance more slowly on the SDSS sample, possibly because of the difficulty of modelling real observational noise. NIHAO UHD deviates in the opposite direction of SDSS from the other NIHAO samples: the importance of its PCs drops to zero much much faster.

##### Now let's take a look at what the first few PCs look like in terms of linear combination of the latent variables

```r
first_few_PCs <- data.frame(sdss_1=prcomp_sdss$rotation[,1], noagn_1=prcomp_nihao_noagn$rotation[,1],
           agn_1=prcomp_nihao_agn$rotation[,1], n80_1=prcomp_nihao_n80$rotation[,1],
           uhd_1=prcomp_nihao_uhd$rotation[,1], sdss_2=prcomp_sdss$rotation[,2], noagn_2=prcomp_nihao_noagn$rotation[,2],
           agn_2=prcomp_nihao_agn$rotation[,2], n80_2=prcomp_nihao_n80$rotation[,2],
           uhd_2=prcomp_nihao_uhd$rotation[,2])

library('lsa')

round(cosine(as.matrix(first_few_PCs)),digits=1)
```

The first principal components of NIHAO AGN, NIHAO No AGN, and NIHAO n80 are all very correlated with each other and with the first PC of SDSS. But NIHAO UHD's first PC does not correlate much with them (the fact that the sign is reversed does not matter). The second principal components of each set of simulations is orthogonal by construction to the first principal component of the same set, so the correlation is 0. Surprisingly though the second principal components do not correlate much with each other across sets. Why is that?
